```python
import networkx as nx
import osmnx as ox
import geopandas
import pandas as pd
import matplotlib.pyplot as plt
from cenpy import products
import numpy as np
from sklearn.cluster import KMeans, AgglomerativeClustering
import seaborn

%matplotlib inline
ox.__version__
```

```python
tags = {"amenity": "post_office"}

place = 'Washington, DC, USA'
G = ox.graph_from_place(place, network_type='drive')

posts = ox.features_from_place(place, tags)

fig, ax = ox.plot_graph(G, bgcolor='white', edge_color='black',
                        node_size=0, edge_linewidth=0.5,
                        show=False, close=False)

posts.plot(ax=ax, color='r', alpha=0.5)

plt.title('Road Network with Post Office Locations in Washington, DC')

yellow_patch = plt.Line2D([0], [0], linestyle='-', color='white', label='Road Network')
white_patch = plt.Circle((0, 0), 0.5, color='red', ec='black', label='Post Offices')
plt.legend(handles=[yellow_patch, white_patch], loc='upper right')

plt.savefig("post_offices_dc.png", format='png', dpi=300)

plt.show()
```
